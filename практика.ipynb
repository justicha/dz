```python
def stations(houses):
    stations = []
    i = 0
    n = len(houses)
    distance = 4
    
    while i < n:
        leftmost = houses[i]
        
        station = leftmost + distance
        stations.append(station)
        
        while i < n and houses[i] <= station + distance:
            i += 1
    
    return stations


houses = [1, 6, 7, 9, 15, 20]
result = stations(houses)
print("Координаты станций:", result)
print("Количество станций:", len(result))
```
