```python
def license_ordering(rates):
    n = len(rates)
    
    licenses = [(rates[j], j) for j in range(n)]
    
    licenses.sort(reverse=True)
    
    order = [idx for _, idx in licenses]
    return order, licenses

rates = [0.9, 0.8, 0.95]
order, sorted_licenses = license_ordering(rates)

print("Оптимальный порядок покупки (индексы лицензий):", order)
print("Коэффициенты в порядке покупки:", [r for r, _ in sorted_licenses])

print('\n')

initial_price = 100
total_cost = 0
for month, (rate, license_idx) in enumerate(sorted_licenses):
    cost = initial_price * (rate ** month)
    total_cost += cost
    print(f"Месяц {month+1}: Лицензия {license_idx+1} (r={rate}) стоит ${cost:.2f}")

print(f"\nОбщая стоимость: ${total_cost:.2f}")
```

```python
def count_subsequence(events, target):
    n = len(events)
    m = len(target)
    
    if m == 0:
        return 0
    
    def find(event_idx, target_idx):
        if target_idx == m:  # Нашли полную подпоследовательность
            return 1
        if event_idx == n:   # Закончились события
            return 0
        
        if events[event_idx] == target[target_idx]:
            # Текущее событие совпадает: можем взять его или пропустить
            return find(event_idx + 1, target_idx + 1) + find(event_idx + 1, target_idx)
        else:
            return find(event_idx + 1, target_idx)
    
    return find(0, 0)


def find_all_pairs(events):
    """Находит все пары (подпоследовательности из 2 событий) и их частоту"""
    pairs = {}
    n = len(events)
    
    for i in range(n):
        for j in range(i + 1, n):
            pair = (events[i], events[j])
            pairs[pair] = pairs.get(pair, 0) + 1
    
    return pairs


def is_subsequence(sequence, target):
    """Проверяет является ли target подпоследовательностью sequence"""
    j = 0
    for i in range(len(sequence)):
        if j < len(target) and sequence[i] == target[j]:
            j += 1
    return j == len(target)


S = ['Yahoo', 'eBay', 'Yahoo', 'Oracle', 'Amazon']

print(f"\nПоследовательность S: {S}\n")

print("Вопрос 1: Все пары событий и их частота:")
pairs = find_all_pairs(S)
for pair, count in sorted(pairs.items(), key=lambda x: -x[1]):
    print(f"  {pair[0]} → {pair[1]}: {count} раз(а)")

print(f"\nВопрос 2: Является ли подпоследовательностью S?")
result = is_subsequence(S, ['Yahoo', 'Oracle'])
print(f"  Ответ: {'ДА' if result else 'НЕТ'}")

result = is_subsequence(S, ['Yahoo', 'eBay', 'Oracle'])
print(f"  Ответ: {'ДА' if result else 'НЕТ'}")

print(f"\nВопрос 3: Сколько раз встречается подпоследовательность ")
count = count_subsequence(S, ['Yahoo', 'Oracle'])
print(f"  Ответ: {count} раз(а)")
```
