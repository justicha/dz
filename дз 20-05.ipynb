```python
def max_profit(p):
    n = len(p)
    if n < 2:
        return None

    min_price = p[0]   
    buy_day = 0   
    best_profit = 0
    best_buy = -1
    best_sell = -1

    for j in range(1, n):
        # Если продать в день j, лучший выбор покупки — min_price
        profit = p[j] - min_price
        if profit > best_profit:
            best_profit = profit
            best_buy = buy_day
            best_sell = j

        # Обновляем минимум для будущих дней
        if p[j] < min_price:
            min_price = p[j]
            buy_day = j

    if best_profit <= 0:
        return None
    return (best_buy, best_sell, best_profit)
```

```python
max_profit([9, 1, 5])
```

```python
print(max_profit([5, 4, 3, 2, 1]))
```

```python
max_profit([1, 2, 3, 4, 5])
```
