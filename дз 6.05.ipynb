```python
def schedule_triathlon(participants):
    sorted_participants = sorted(
        participants,
        key=lambda p: -(p["bike"] + p["run"])
    )
    
    pool_free_time = 0
    finish_times = []
    
    print("Порядок выхода участников:")
    for p in sorted_participants:
        start_swim = pool_free_time
        finish_swim = start_swim + p["swim"]
        finish_bike = finish_swim + p["bike"]
        finish_run = finish_bike + p["run"]
        
        finish_times.append(finish_run)
        pool_free_time = finish_swim
        
        print(f"  {p['name']}: старт плавания {start_swim} → финиш {finish_run}")
    
    total_time = max(finish_times)
    print(f"\nОбщее время соревнования: {total_time}")
    
    return total_time

participants = [
    {"name": "Алиса", "swim": 20, "bike": 10, "run": 3},
    {"name": "Боб", "swim": 15, "bike": 8, "run": 2},
    {"name": "Виктор", "swim": 25, "bike": 12, "run": 4},
    {"name": "Галина", "swim": 18, "bike": 9, "run": 2},
]

schedule_triathlon(participants)
```
