```python
def schedule_triathlon(participants):
    sorted_participants = sorted(
        participants, 
        key=lambda p: p["swim"] + p["bike"] + p["run"]
    )
    
    for i, p in enumerate(sorted_participants, 1):
        total_time = p["swim"] + p["bike"] + p["run"]
        print(f"{i}. {p['name']} - общее время: {total_time} мин")
    
    return sorted_participants


participants = [
    {"name": "Алиса", "swim": 20, "bike": 10, "run": 3},
    {"name": "Боб", "swim": 15, "bike": 8, "run": 2},
    {"name": "Виктор", "swim": 25, "bike": 12, "run": 4},
    {"name": "Галина", "swim": 18, "bike": 9, "run": 2},
]

result = schedule_triathlon(participants)
```
