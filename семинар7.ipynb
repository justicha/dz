```python
def check_consistency(n, evaluations):
    graph = {i: [] for i in range(1, n + 1)}
    for i, j, kind in evaluations:
        graph[i].append((j, kind))
        graph[j].append((i, kind))

    color = [-1] * (n + 1)

    for start in range(1, n + 1):
        if color[start] != -1:
            continue

        color[start] = 0
        queue = [start]
        index = 0

        while index < len(queue):
            u = queue[index]  
            index += 1      

            for v, kind in graph[u]:
                if kind == "same":
                    needed = color[u]
                else:
                    needed = 1 - color[u]

                if color[v] == -1:
                    color[v] = needed
                    queue.append(v)
                else:
                    if color[v] != needed:
                        print("ПРОТИВОРЕЧИЕ")
                        return

    species = {}
    for i in range(1, n + 1):
        species[i] = "A" if color[i] == 0 else "B"
    print("НЕПРОТИВОРЕЧИВО")
    for sample, sp in species.items():
        print(f"  Образец {sample} → вид {sp}")
```
